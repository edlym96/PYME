### Some code for visualizing and printing the board

```python
# Helper function to print frames
def print_frame(row: int):
    if (row % 3) == 0:
        print("  +===========+===========+===========+")

    else:
        print("  +---+---+---+---+---+---+---+---+---+")

# Function to print board
def print_board(input_array):
    # Check if input is a 2d array and square
    if not all(isinstance(el, list) for el in input_array) or any(len(el) != len(input_array) for el in input_array):
        raise TypeError

    print(end=" ")
    for i in range(1,10):
        print("   " + str(i), end = "")
    print("")
    for index, row in enumerate(input_array):
        print_frame(index)
        print(chr(ord('A')+index), end="")
        print(" |", end="")
        for col in row:
            # Check if entry is a digit or a '.'
            if (not str(col).isdigit()) and col != '.'  :
                raise ValueError
            else:
                print("  " if col == "." else " "+str(col), end=" |")
        print("")
    print("  +===========+===========+===========+")
```
